```python
# Must read with "/" not "\"
BLOCK_PATH = 'C:/Users/yongc/Desktop/Synapse/Tanks/VirusTest-240116-082958/dLight_mPFC-240209-121108'
data = tdt.read_block(BLOCK_PATH)
```

```python
data = tdt.read_block(BLOCK_PATH, evtype=['streams', 'epocs'], channel=1)
print(data.streams)
```

```python
time_405A = np.linspace(1, len(data.streams._405A.data), len(data.streams._405A.data)) / data.streams._405A.fs
time_465A = np.linspace(1, len(data.streams._465A.data), len(data.streams._465A.data)) / data.streams._465A.fs
```

```python
data.streams._465A.data*1e6
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(12, 8), sharex=True)

ax1.set_ylim([150*1e6, 200*1e6])
ax1.plot(time_405A, data.streams._405A.data*1e6, color='cornflowerblue')
ax1.set_title('Basic Data Plotting: \n405 Waveform', fontsize=14)

ax2.set_ylim([150*1e6, 200*1e6])
ax2.plot(time_465A, data.streams._465A.data*1e6, color='firebrick')
ax2.set_title('465 Waveform', fontsize=14)

#ax2.set_xlabel('Time (s)', fontsize=12)
#x2.set_ylabel('Amplitude ($\mu$V)', fontsize=12)
#ax1.set_ylabel('Amplitude ($\mu$V)', fontsize=12)
ax1.set_xlim(6, 409.8)
ax2.set_xlabel('Time (s)', fontsize=12)

plt.show()
```
