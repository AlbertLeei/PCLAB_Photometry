#### Three Methods to access wave store sampling rate
1. Dot Syntax: data.streams._405A.fs   
2. Dict Keys Only: data['streams']['Wav1']['fs']
3. Mix of Dot adn Dict: data['streams']._405A['fs']

```python
print(data.streams._465A)
```

```python
num_samples = len(data.streams._465A.data)
samp_rate = data.streams._465A.fs

print('channel 1:', data.streams._465A.data)
print('number of samples:', num_samples)
print('Sampling Rate (fs):', samp_rate)
```

Create time vector for plotting by dividing the number of samples in the array by the sampling rate

```python
_465_time = np.linspace(1, num_samples, num_samples) / samp_rate
```

Plotting first 2 seconds from a single channel

```python
data.streams._465A.data[1]
```

```python
t = int(2 * samp_rate) # int rounds it to the nearest integer

# declare the figure size
fig1 = plt.subplots(figsize=(10, 6))

channel = 1 

# plot the line using slices
plt.plot(_465_time[0:t], data.streams._465A.data[0:t], color='cornflowerblue')

# Some matplotlib stuff
# add an annotation mark to the figure
plt.annotate('Point Here',
             xy=(0.8,0.002),
             xytext=(.88,.003),
             arrowprops=dict(arrowstyle='->', color='k')
            )

# create title, axis labels, and legend
plt.title('_465 Data', fontsize=16)
plt.xlabel('Seconds', fontsize=14)
plt.ylabel('Volts', fontsize=14)
plt.legend(('Channel {}'.format(channel),),
           loc='lower right', 
           bbox_to_anchor=(1.0,1.01)
          )
plt.autoscale(tight=True)
plt.show()
```
